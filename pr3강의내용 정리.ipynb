텍스트 마이닝이란
    
    - 비정형 txt data를 정형화 하는것, 즉 pattern or trend로 나타내는것
    
    - 가변으로 되있는 문서를 일정한 길이의 vector로 변환한다: 이것은 숫자 여야함
    
    - text classification 단어 분류 범위 안에 sentiment analysis : 감성분석(어떤글이 우호적인지 아닌지, spam이냐 아니냐)이 들어있다
    
    - document summariazation 글 내용 요약
    
    - translation 번역
    
    - prediction 예측 등에 사용
    
    -스크레이핑과 크롤링방식이 있다. 스크레이핑은 웹사이트에서 특정 정보 추출하는 기술이고 html형식을 '가공'해서 사용한다.
    
    -크롤링은 웹사이트에서 정기적으로 정보를 추출하는 기술이다.
    
    - selenium을 사용하는 방식이 있고, urllib, beautifulsoup, css selector 등을 사용해서 웹을 크롤링한다.
    
    - 기본도구는 nlp라고 하며, nltk가 가장많이 알려진 nlp 라이브러리 - 영어기준

```python
from bs4 import BeautifulSoup
from urllib import request
```

```python
url = "https://ko.wikipedia.org/wiki/%EC%95%84%EC%A3%BC%EB%8C%80%ED%95%99%EA%B5%90#%EC%9D%B4%EB%85%90"

res = request.urlopen(url)
soup = BeautifulSoup(res, "html.parser")

selector = "#bodyContent"
#아주대학교 정보 위키에서 크롤링 해보기


for a in soup.select(selector):
    title = a.text
    print(title)
```
